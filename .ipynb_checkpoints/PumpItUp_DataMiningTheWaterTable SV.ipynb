Hypothesis test: Both the values are not correlated of each other

Alternate Hypothesis: Both the values are correlated of each other

If the p-value here is less than 0.05. Therefore, we will reject our null hypothesis and accept alternate hypothesis

### Imputing None and Unknown values with Mode of their categorical values

```python
fullcorpus.scheme_management.unique()
```

```python
# Imputing None woith "vwc" for scheme_management
data_m['scheme_management']=data_m['scheme_management'].replace('None','VWC')
# Imputing nan with "vwc" for scheme_management
data_m['scheme_management']=data_m['scheme_management'].fillna('VWC')

# Imputing unknown woith "vwc" for management 
data_m['management']=data_m['management'].replace('unknown','VWC')

# Imputing unknown woith "user-grou" for management_group 
data_m['management_group']=data_m['management_group'].replace('unknown','user-group')

# Imputing unknown woith "never pay" for payment 
data_m['payment']=data_m['payment'].replace('unknown','never pay')

# Imputing unknown woith "never pay" for payment_type 
data_m['payment_type']=data_m['payment_type'].replace('unknown','never pay')

# Imputing unknown woith "never pay" for quantity 
data_m['quantity']=data_m['quantity'].replace('unknown','enough')

# Imputing unknown woith "never pay" for source 
data_m['source']=data_m['source'].replace('unknown','spring')

# Imputing unknown woith "never pay" for source_class 
data_m['source_class']=data_m['source_class'].replace('unknown','groundwater')
```

Construction year can never be 0. we are imputing mode in the place of Zeros

```python
mode_value = data_m.construction_year[data_m.construction_year!=0].mode()
data_m['construction_year']=data_m['construction_year'].replace(0,mode_value[0])

data_m['construction_year'].hist()

```

```python
# Getting the age of the pump

def date_to_diff2(x):
    if x == 0:
        return 0
    else:
        return 2020 - x
    


data_m['years_cons'] = data_m['construction_year'].apply(date_to_diff2)

data_m['years_cons'].hist()
```

```python
data_m['years_cons_bin'] = pd.cut(data_m['years_cons'], [0, 20, 40,60], labels=['0-20', '20-40','>40'], right=False)

# We created gps_height_bin using gps_height, so droppinggps_height column

data_m.drop(['years_cons','construction_year'],axis=1,inplace = True)
```

### Creating bins for gps_height matching to above distribution

```python
data_m['gps_height_bin'] = pd.cut(data_m['gps_height'], [-100,0, 500, 1000,1500,2000,2500,3000], labels=['<0','0-500', '500-1000', '1000-1500','1500-2000','2000-2500','2500-3000'], right=False)

# We created gps_height_bin using gps_height, so droppinggps_height column

data_m.drop('gps_height',axis=1,inplace = True)
```
